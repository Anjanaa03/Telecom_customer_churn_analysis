```python
df = df = pd.read_csv('Customer Churn.csv')
def conv(value):
   if value == 1 :
       return "yes"
   else :
       return "no"
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
df
```

```python
#convert the 0 or 1 values to yes or no 
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0]) #for labeling data
plt.title("Count of Customers by Churn")
plt.show() 
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
#groupby and find out the churn by counting string
plt.pie(gb['Churn'] , labels = gb.index , autopct ="%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
gb
```

```python
# from the given pie chart we can understand 26.54% of ur customer have churned out 
# now lets explore the reason behind it 
```

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender", data = df, hue = "Churn") 
plt.show()
```

```python
plt.figure(figsize= (3,3))
sns.countplot(x = "SeniorCitizen", data = df, hue="Churn") 
plt.title("Churn by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize= (3,3))
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert counts to percentages
ct_percent = ct.div(ct.sum(1), axis=0) * 100

# Plot stacked bar chart
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(6, 4))

# Add percentage labels on the bars
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    if height > 0:
        ax.text(x + width/2, y + height/2, f'{height:.1f}%', 
                ha='center', va='center', color='white', fontsize=10)

# Add title and labels
plt.title('Churn by Senior Citizen (Percentage)')
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')

# Show plot
plt.show()
```

```python
#comparitively a greater percent of senior citizens have churned out
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure" , data = df , bins = 72 , hue = 'Churn')
plt.title("Histogram of tenure")
# we chose 72 bcz thats the max tenure
plt.show() 
```

```python
#people who used services for a long time stayed and people who used it for few months
# churned out more , we can say that something is forcing them to churn out initially 
# or people didnt churn out after staying for a long time because they were satisfied 
```
